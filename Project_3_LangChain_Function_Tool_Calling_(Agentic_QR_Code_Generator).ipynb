```python
################################################################
#                    HELPER: EXTENSION UNIFICATION             #
################################################################

def unify_extension(
    filename: str,
    default_ext=".png",
    supported={".png", ".jpg", ".jpeg", ".bmp", ".gif"}
) -> str:
    """
    Ensures the file extension is one of the supported types.
    If not provided or unsupported, uses default_ext (e.g., '.png').
    """
    base, ext = os.path.splitext(filename)
    ext_lower = ext.lower()

    # If user typed no extension
    if not ext_lower:
        return base + default_ext

    # If user typed an extension that isn't in 'supported'
    if ext_lower not in supported:
        return base + default_ext

    # unify to correct extension in lower
    for s in supported:
        if s == ext_lower:
            return base + s
    return base + default_ext

################################################################
#          YES/NO CHECK: FLEXIBLE INPUT INTERPRETATION         #
################################################################

def is_yes(answer: str) -> bool:
    """
    Returns True if the user typed an affirmative response
    (e.g., 'y', 'yes', 'YES', 'Yes'),
    otherwise False.
    """
    return answer.strip().lower() in ["y", "yes"]

def is_no(answer: str) -> bool:
    """
    Returns True if the user typed a negative response
    (e.g., 'n', 'no', 'NO', 'No'),
    otherwise False.
    """
    return answer.strip().lower() in ["n", "no"]

################################################################
#       DEFINING QR CODE GENERATION FUNCTION & DATA MODEL      #
################################################################

class QRCodeInput(BaseModel):
    data: str = Field(..., description="Text, URL, or local file path to encode in the QR code.")
    filename: str = Field("qrcode.png", description="Desired output filename (e.g., 'mycode.png').")

def generate_qr_code(input: QRCodeInput = None, **kwargs) -> str:
    """
    Generates a QR code image from the input data and saves it locally.

    Args:
        input (QRCodeInput):
          data: text/URL/local path to embed in the QR
          filename: desired output file name
        **kwargs:
          fallback for older usage

    Returns:
        str: success message

    Raises:
        ValueError: if data is empty
        RuntimeError: if there's an error saving
    """
    if input:
        data = input.data
        filename = input.filename
    else:
        data = kwargs.get("data")
        filename = kwargs.get("filename", "qrcode.png")

    if not data:
        raise ValueError("Input data must be a non-empty string.")

    # unify extension for multiple formats
    filename = unify_extension(filename)

    try:
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )
        qr.add_data(data)
        qr.make(fit=True)
        img = qr.make_image(fill_color="black", back_color="white")
        img.save(filename)
        return f"QR code successfully generated and saved as '{filename}'."
    except Exception as e:
        raise RuntimeError(f"Failed to generate or save QR code: {e}")
```



#### **Helper: Extension Unification**
1. **`def unify_extension(...)`**  
   - **Explanation**:  
     📁 Ensures a given filename has a valid and supported extension. If the user provides no extension or an unsupported one, it appends a default extension (e.g., `.png`).  
   - **Use**:  
     📁 Handles filenames flexibly, ensuring compatibility with supported formats.

2. **`os.path.splitext(...)`**  
   - **Explanation**:  
     🖥️ Splits the filename into its base name and extension (e.g., `"file.png"` becomes `("file", ".png")`).  
   - **Use**:  
     🖥️ Used to extract or manipulate file extensions.

3. **`if not ext_lower:`**  
   - **Explanation**:  
     📌 Checks if the user didn't provide an extension.  
   - **Use**:  
     📌 Appends the default extension when no extension is provided.

4. **`if ext_lower not in supported:`**  
   - **Explanation**:  
     🚫 Ensures the given extension is among the supported ones. If not, it replaces it with the default.  
   - **Use**:  
     🚫 Prevents unsupported formats from causing errors.

```python
################################################################
#                  CONFIGURE LLM (GEMINI)                      #
################################################################

# Use your key from Google AI Studio or wherever you have it
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')  # Or manually: os.environ["GOOGLE_API_KEY"] = "YOUR_KEY_HERE"

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-exp", temperature=0)

# Create the structured tool for QR code generation
qr_code_tool = StructuredTool.from_function(
    func=generate_qr_code,
    name="generate_qr_code",
    description="Generates a QR code from the provided data and saves it as an image file.",
    args_schema=QRCodeInput
)

# Build an agent with minimal logging
agent = initialize_agent(
    tools=[qr_code_tool],
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)
```

```python
################################################################
#                        MAIN SCRIPT                           #
################################################################

# 1) Ask user in Markdown if they want to upload from local
display(Markdown("### Do you want to upload a file from your computer? (yes/no)"))
user_input = input().strip()

if is_yes(user_input):
    display(Markdown("**Please select a file to upload from your local machine:**"))
    uploaded_files = files.upload()
    if uploaded_files:
        first_uploaded = list(uploaded_files.keys())[0]
        user_data = first_uploaded
        display(Markdown(f"**You uploaded:** `{first_uploaded}`"))
    else:
        display(Markdown("_No file uploaded. Provide a text/URL/local path manually._"))
        display(Markdown("**Enter the text/URL/local path to embed in QR code:**"))
        user_data = input()
elif is_no(user_input):
    # if user explicitly said no
    display(Markdown("**Enter the text/URL/local path to embed in QR code:**"))
    user_data = input()
else:
    # unrecognized => treat as no
    display(Markdown("_Unrecognized response, defaulting to NO upload._"))
    display(Markdown("**Enter the text/URL/local path to embed in QR code:**"))
    user_data = input()

# 2) Ask for the desired output filename in Markdown
display(Markdown("**Enter desired output filename (e.g. 'myqr.jpg' or 'image.png'):**"))
user_filename = input().strip()
if not user_filename:
    user_filename = "myqr.png"



# 3) Build the LLM query
query = f"Generate a QR code for the URL '{user_data}' and save it as '{user_filename}'."

# 4) Invoke the agent
response = agent.invoke(query)

# 5) Show the final response in Markdown
display(Markdown(f"**LLM's Final Response:** {response}"))

# 6) Attempt to unify extension & download the file
final_filename = unify_extension(user_filename)
try:
    files.download(final_filename)
except FileNotFoundError:
    display(Markdown(f"**File '{final_filename}' not found.** Verify if the QR code was generated."))
```
