```python
################################################################
#                    HELPER: EXTENSION UNIFICATION             #
################################################################

def unify_extension(
    filename: str,
    default_ext=".png",
    supported={".png", ".jpg", ".jpeg", ".bmp", ".gif"}
) -> str:
    """
    Ensures the file extension is one of the supported types.
    If not provided or unsupported, uses default_ext (e.g., '.png').
    """
    base, ext = os.path.splitext(filename)
    ext_lower = ext.lower()

    # If user typed no extension
    if not ext_lower:
        return base + default_ext

    # If user typed an extension that isn't in 'supported'
    if ext_lower not in supported:
        return base + default_ext

    # unify to correct extension in lower
    for s in supported:
        if s == ext_lower:
            return base + s
    return base + default_ext

################################################################
#          YES/NO CHECK: FLEXIBLE INPUT INTERPRETATION         #
################################################################

def is_yes(answer: str) -> bool:
    """
    Returns True if the user typed an affirmative response
    (e.g., 'y', 'yes', 'YES', 'Yes'),
    otherwise False.
    """
    return answer.strip().lower() in ["y", "yes"]

def is_no(answer: str) -> bool:
    """
    Returns True if the user typed a negative response
    (e.g., 'n', 'no', 'NO', 'No'),
    otherwise False.
    """
    return answer.strip().lower() in ["n", "no"]

################################################################
#       DEFINING QR CODE GENERATION FUNCTION & DATA MODEL      #
################################################################

class QRCodeInput(BaseModel):
    data: str = Field(..., description="Text, URL, or local file path to encode in the QR code.")
    filename: str = Field("qrcode.png", description="Desired output filename (e.g., 'mycode.png').")

def generate_qr_code(input: QRCodeInput = None, **kwargs) -> str:
    """
    Generates a QR code image from the input data and saves it locally.

    Args:
        input (QRCodeInput):
          data: text/URL/local path to embed in the QR
          filename: desired output file name
        **kwargs:
          fallback for older usage

    Returns:
        str: success message

    Raises:
        ValueError: if data is empty
        RuntimeError: if there's an error saving
    """
    if input:
        data = input.data
        filename = input.filename
    else:
        data = kwargs.get("data")
        filename = kwargs.get("filename", "qrcode.png")

    if not data:
        raise ValueError("Input data must be a non-empty string.")

    # unify extension for multiple formats
    filename = unify_extension(filename)

    try:
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )
        qr.add_data(data)
        qr.make(fit=True)
        img = qr.make_image(fill_color="black", back_color="white")
        img.save(filename)
        return f"QR code successfully generated and saved as '{filename}'."
    except Exception as e:
        raise RuntimeError(f"Failed to generate or save QR code: {e}")
```

```python
################################################################
#                  CONFIGURE LLM (GEMINI)                      #
################################################################

# Use your key from Google AI Studio or wherever you have it
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')  # Or manually: os.environ["GOOGLE_API_KEY"] = "YOUR_KEY_HERE"

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-exp", temperature=0)

# Create the structured tool for QR code generation
qr_code_tool = StructuredTool.from_function(
    func=generate_qr_code,
    name="generate_qr_code",
    description="Generates a QR code from the provided data and saves it as an image file.",
    args_schema=QRCodeInput
)

# Build an agent with minimal logging
agent = initialize_agent(
    tools=[qr_code_tool],
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)
```
